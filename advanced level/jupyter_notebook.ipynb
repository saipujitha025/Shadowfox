```python
#Research Question
"""What are the key factors that influence customer churn in the banking sector, and how accurately can we predict churn using machine learning models?
The factors that effect the customer churn in banking sector are gender, age group, tenure, credit score and balance.
```

```python
#churn rate by gender
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Exited', data=df, palette='viridis')
plt.title('Churn Rate by Gender')
plt.show()
```

```python
#churn rate by age group
plt.figure(figsize=(10,5))
df['AgeGroup']=pd.cut(df['Age'],bins=[18,30,40,50,60,100], labels=['18-30', '30-40','40-50','50-60','60+'])
sns.countplot(x='AgeGroup', hue='Exited', data=df, palette='magma')
plt.title('Churn Rate by Age Group')
plt.show()
```

```python
#Churn rate by tenure
plt.figure(figsize=(10,5))
sns.boxplot(x='Exited', y='Tenure', data=df, hue='Exited' , palette='coolwarm')
plt.title('Churn Rate by Tenure')
plt.show()
```

```python
#Churn by credit score
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='CreditScore', hue='Exited', kde=True, palette='pastel', bins=30)
plt.title('Churn Distribution by Credit Score')
plt.show()

```

```python
#Churn rate by balance
plt.figure(figsize=(10,5))
sns.boxplot(x='Exited', y='Balance',hue='Exited', data=df, palette='Set2')
plt.title('Churn rate by Account Balance')
plt.show()
```
